# Importing necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.utils import resample
```

# Loading and exploring the dataset

```python
mcdonalds = pd.read_csv("https://homepage.boku.ac.at/leisch/MSA/datasets/mcdonalds.csv")
```

```python

print(mcdonalds.columns)
print(mcdonalds.shape)
print(mcdonalds.head())
```

# preprocessing the data

```python

MD_x = (mcdonalds.iloc[:, :11] == "Yes").astype(int)
print(MD_x.mean().round(2))
```

# performing PCA

```python
pca = PCA()
MD_pca = pca.fit_transform(MD_x)
print(pca.explained_variance_ratio_)
print(pd.DataFrame(MD_pca).head())
```

```python
plt.scatter(MD_pca[:, 0], MD_pca[:, 1], color='grey')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of McDonald\'s Data')
plt.show()
```

# extracting Segments using k-mean

```python
np.random.seed(1234)
inertia = []
for k in range(2, 9):
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(MD_x)
    inertia.append(kmeans.inertia_)
plt.plot(range(2, 9), inertia)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

# Checking the Stability of the cluster

```python
kmeans = KMeans(n_clusters=4, n_init=10, random_state=1234)
mcdonalds['k4'] = kmeans.fit_predict((mcdonalds.iloc[:, :11] == "Yes").astype(int))
```

```python

from sklearn.metrics import silhouette_score

silhouette_scores = []

cluster_range = range(2, 9)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=1234)
    labels = kmeans.fit_predict(MD_x)
    silhouette_scores.append(silhouette_score(MD_x, labels))
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Cluster Stability (Silhouette Score)')
plt.grid(True)
plt.show()
```

# Clustering and Profilling Segments

```python
kmeans = KMeans(n_clusters=4, n_init=10, random_state=1234)
MD_k4 = kmeans.fit_predict(MD_x)
print(pd.Series(MD_k4).value_counts())
```

# Hierarchical clustering

```python
linkage_matrix = linkage(MD_x.T, method='ward')
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Feature')
plt.ylabel('Distance')
plt.show()
```

# Mixture of distributions using Gaussian Mixture Models

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=4, n_init=10, random_state=1234)
MD_gmm = gmm.fit_predict(MD_x)
print(pd.crosstab(MD_k4, MD_gmm))
```

# Define the mapping from categorical to numeric values

```python

like_mapping = {
    'I hate it!-5': 1,
    'I love it!+5': 5
}

mcdonalds['Like_n'] = mcdonalds['Like'].map(like_mapping)

formula = 'Like_n ~ ' + ' + '.join(mcdonalds.columns[:11])
md_reg = ols(formula, data=mcdonalds).fit()
print(md_reg.summary())
```

# Profiling segments

```python
boxplot_data = [mcdonalds[mcdonalds['k4'] == i]['Age'] for i in range(4)]
plt.boxplot(boxplot_data, notch=True, labels=range(4))
plt.xlabel('Segment Number')
plt.ylabel('Age')
plt.title('Age Distribution by Segment')
plt.show()
```

# Encoding Part

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
mcdonalds['VisitFrequency_encoded'] = label_encoder.fit_transform(mcdonalds['VisitFrequency'])
print(mcdonalds['VisitFrequency_encoded'])
```

```python
label_encoder = LabelEncoder()
mcdonalds['Gender_encoded'] = label_encoder.fit_transform(mcdonalds['Gender'])
print(mcdonalds['Gender_encoded'])
```

```python
mcdonalds.head()
```

#  Decision tree for segment profiling

```python
X = mcdonalds[['Like_n', 'Age', 'VisitFrequency_encoded', 'Gender_encoded']]
y = (MD_k4 == 3).astype(int)
clf = DecisionTreeClassifier(random_state=1234)
clf.fit(X, y)
feature_names = X.columns.tolist()

plt.figure(figsize=(15, 10))
plot_tree(clf, feature_names=feature_names, class_names=['Other', 'Segment 3'], filled=True)
plt.show()
```

# Describing segments

```python
visit = mcdonalds.groupby('k4')['VisitFrequency_encoded'].mean()
like = mcdonalds.groupby('k4')['Like_n'].mean()
female = mcdonalds.groupby('k4')['Gender_encoded'].apply(lambda x: (x == 'Female').mean())

plt.scatter(visit, like, s=female*1000, alpha=0.5)
for i in range(len(visit)):
    plt.text(visit.iloc[i], like.iloc[i], str(i))
plt.xlabel('Visit Frequency')
plt.ylabel('Like')
plt.title('Segment Description')
plt.show()
```
